# Training Classifiers

```
X = df[['mass', 'width', 'height', 'color_score']]
y = df['fruit_label']
```

```
X.columns
```

```
from sklearn.tree import DecisionTreeClassifier, plot_tree

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X,y)
```

## GINI - no split

```
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
tree = plot_tree(clf, feature_names = ['mass', 'width', 'height', 'color_score'], class_names=['apple','mandarin','orange','lemon'] , filled=True) 
print(tree)
```

## GAIN(entopy) - no split

```
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X,y)
```

```
plt.figure(figsize=(15, 15))
tree = plot_tree(clf, feature_names = ['mass', 'width', 'height', 'color_score'], class_names=['apple','mandarin','orange','lemon'] , filled=True) 
print(tree)
```

## GINI - 80/20 split

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```
clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X_train,y_train)
plt.figure(figsize=(10, 10))
tree = plot_tree(clf, feature_names = ['mass', 'width', 'height', 'color_score'], class_names=['apple','mandarin','orange','lemon'] , filled=True) 
print(tree)
```

```
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred, normalize=False)
```

```
clf.score(X_test,y_test)
```

### Precision, Recall, Fscore, Support

```
from  sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average=None) # labels=['apple','mandarin','orange','lemon'] 
```
