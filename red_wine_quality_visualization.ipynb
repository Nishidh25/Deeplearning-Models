## Histograms

A graphical display of data using bars of different heights. 

It is similar to a Bar Chart, but a histogram groups numbers into ranges . 

The height of each bar shows how many fall into each range.

```
df["fixed.acidity"].hist(bins=20)
```

```
df["chlorides"].hist(bins=20)
```

```
df["total.sulfur.dioxide"].hist(bins=20)
```

**Analysis** of above 3 Histograms : Right Skewed Distributions

Here Mean > Mode

Data skewed to the right is usually a result of a lower boundary in the selected data set attributes 

```
df["pH"].hist(bins=20)
```

**Analysis** of the above Histogram: Normal Distribution 

Mean ~ Mode

In a normal distribution, points are as likely to occur on one side of the average as on the other

## Pair Plot

Pair Plots are a really simple way to visualize relationships between each variable

```
import seaborn as sns
sns.pairplot(df,hue='quality');
```

**Analysis**: 

The pair plot contains scatter plots between every attribute in the dataframe 

Every scatterplot should always include a description of 
* form - linear or nonlinear association
* direction - positive or negative association 
* strength of the association - strong, moderately strong, or weak
* any outliers? 


Analysing few the scatter plots above:

```
sns.scatterplot(x='density', y='fixed.acidity',hue='quality', data=df)
```

In the above scatter plot,

## Correlation matrix heatmap

A correlation matrix is a table showing correlation coefficients between variables. 

Each cell in the table shows the correlation between two variables.

A correlation matrix is used to summarize data, as an input into a more advanced analysis, and as a diagnostic for advanced analyses.

```
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
```

**Analysis of correlation matrix**:

Correlation Coefficients
* +1: A perfect positive relationship
* 0.8: A fairly strong positive relationship
* 0.6: A moderate positive relationship
* 0.2: Weak positive relationship
* 0: No relationship. As one value increases, there is no tendency for the other value to change in a specific direction
* -0.2: Weak negative relationship
* -0.6: A moderate negative relationship
* -0.8: A fairly strong negative relationship
* -1: A perfect negative relationship


We can use correlation values [-1,-0.3] and [0.3,1] which will affect the prediction made, as (-0.3,0.3) range will not affect(or slightly affect or might overfit) the predictions since those variables would have weak relationship.

There is no one-size fits all best answer for how strong a relationship should be. The correct correlation value depends on the data.


In the above matrix , the highest corellation(apart from self attribues) is 0.67 and -0.68

## Parallel Coordinates plot

Parallel coordinates are a common way of visualizing high-dimensional geometry and analyzing multivariate data.

 To show a set of points in an n-dimensional space, a backdrop is drawn consisting of n parallel lines, typically vertical and equally spaced.

```
plt.figure(figsize=(20, 10))
pd.plotting.parallel_coordinates(df, "quality")
```

**Analysis**

The values in a parallel coordinate plot are always normalized. This means that for each point along the X-axis, the lowest value in the corresponding column is set to 0% and the highest value in that column is set to 100% along the Y-axis. 

We can find out the exact values of each attributes needed for any quality using this plot 

eg, We can find out all the attributes values required for quality 7 by using this plot  

## Box Plot

A box plot or boxplot is a method for graphically depicting groups of numerical data through their quartiles.

Box plots may also have lines extending from the boxes indicating variability outside the upper and lower quartiles.

```
plt.figure(figsize=(20, 10))
sns.boxplot(data=df)
```

**Analysis:**

Every box plot divide the columns data into sections that each contain approximately 25% of the data in that set.

We can see many outliers , and the range of data in different columns of the dataset

## Contour Plot 
Contour plots (sometimes called Level Plots) are a way to show a three-dimensional surface on a two-dimensional plane.

It graphs two predictor variables X Y on the y-axis and a response variable Z as contours. 
 
These contours are sometimes called z-slices or iso-response values.

A contour plot contains the following elements:
* Predictors on the x- and y-axes.
* Contour lines that connect points that have the same response value
* Colored contour bands that represent ranges of the response values


```
sns.jointplot("volatile.acidity", "citric.acid", data=df, kind="kde",color='b')
```

**Analysis**



This contour plot shows the relationship between volatile.acidity and citric.acid used in wines.

Darker regions indicate high density of data in that range.
